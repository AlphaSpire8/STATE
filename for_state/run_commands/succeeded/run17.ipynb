## 2.finetune

每次需要调整的部分是：

- --output_lora

```python
export CUDA_VISIBLE_DEVICES=4
python /data3/fanpeishan/state/for_state/scripts/finetune_v2.py \
  --model_dir /data3/fanpeishan/state/for_state/models/ST-Tahoe \
  --checkpoint /data3/fanpeishan/state/for_state/models/ST-Tahoe/final_from_preprint.ckpt \
  --adata /data3/fanpeishan/state/for_state/run_results/run15/c37_prep.h5ad \
  --pert_col drugname_drugconc \
  --batch_col plate \
  --output_lora /data3/fanpeishan/state/for_state/run_results/run17/lora_state.pth \
  --save_dir /data3/fanpeishan/state/for_state/run_results/run17/lora_out \
  --epochs 5 \
  --batch_size 128 \
  --lr 2e-4 \
  --lora_rank 16 \
  --lora_alpha 32 \
  --use_delta_loss \
  --lambda_delta 1.0 \
  --lambda_pearson 0.4 \
  --lambda_mse 0.2 \
  --weight_decay 1e-4 \
  --target_modules pert_encoder,basal_encoder,transformer_backbone
```

## 4.cell-eval run

```python
export CUDA_VISIBLE_DEVICES=4
cell-eval run \
    -ap /data3/fanpeishan/state/for_state/run_results/run17/c37_infer.h5ad \
    -ar /data3/fanpeishan/state/for_state/run_results/run15/c37_prep.h5ad \
    -o /data3/fanpeishan/state/for_state/run_results/run17/run_results \
    --control-pert "[('DMSO_TF', 0.0, 'uM')]" \
    --pert-col 'drugname_drugconc' \
    --profile minimal \
    --batch-size 1024 \
    --num-threads 12 
```
